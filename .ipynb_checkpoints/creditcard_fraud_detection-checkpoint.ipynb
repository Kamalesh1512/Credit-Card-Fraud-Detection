1.   Legit transactions - 99.83% (283253 records)
2.   Fraudulent transactions -0.17% (473 records)

## *Amount column summary*

1.   The average value of Amount of a transactions in fraudulent transactions:  $123.87 

2.   The average value of Amount of a transactions in Legit transactions:  $88.41


#### Greater the value of Amount there is a chance of fraud/Abnormal transactions to happen
 


## *Correlation Matrix*

In the credit card fraud detection dataset,

1.  if two features are highly positively correlated, then they may provide redundant information for predicting fraud, and one of them could potentially be removed from the dataset without losing much predictive power. 
2. Conversely, if two features are highly negatively correlated they may be useful for predicting fraud in combination, as changes in one feature could cancel out changes in the other feature.

#### In our dataset No two columns are highly correlated

## Handling Outliers


```python
def outliers(dataset):
    out_dict=dict()
    num_cols=dataset.select_dtypes([float,int]).columns.to_list()
    for column in num_cols[:-1]:
        q1=np.percentile(dataset[column],25)
        q3=np.percentile(dataset[column],75)
        IQR=q3-q1
        low=q1 - (1.5*IQR)
        high=q3 + (1.5*IQR)
        outliers=[value for value in dataset[column] if (value>high) or (value<low)]
        out_dict[column]=outliers
    return out_dict
```

## Handling Imbalanced Data

### 1.Under-Sampling Method to Balance the data  

```python
## Features
X=data.iloc[:,:-1]

#Labels
y=data['Class']
```

```python
## Observation
print("Shape of legit data(Majority Class)",data.shape)
print("Shape of fraud data(Minority Class)",fraud.shape)
print()

under_sample=RandomUnderSampler(random_state=22)
X_Under,y_under=under_sample.fit_resample(X,y)
print("After Under Sampling shape of data:",X_Under.shape)
```

### 2. Over-Sampling Method Using SMOTE

```python

smote=SMOTE(sampling_strategy='minority')
X_smote,y_smote=smote.fit_resample(X,y)

print("shape of data, After Over Sampling(SMOTE) :",X_smote.shape)
```
